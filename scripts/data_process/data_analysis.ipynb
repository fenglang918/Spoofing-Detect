# merge_order_trade

```python
# -*- coding: utf-8 -*-
"""
merge_order_trade.py
生成: 委托事件流.csv
-------------------------------------------------------
逐笔委托.csv   主键: 交易所委托号
逐笔成交.csv   主键: 买/卖序号 (叫买序号 / 叫卖序号)
-------------------------------------------------------
将 data_folder 设置为：'/home/ma-user/code/fenglang/Spoofing Detect/data/exmaple'
"""

import pandas as pd
import numpy as np
from pathlib import Path

# ------------ 0. 数据文件夹 & 文件名配置 ---------------
DATA_FOLDER = Path('/home/ma-user/code/fenglang/Spoofing Detect/data/exmaple')
FILE_ORDER = DATA_FOLDER / '逐笔委托.csv'
FILE_TRADE = DATA_FOLDER / '逐笔成交.csv'
OUTPUT_FILE = DATA_FOLDER / '委托事件流.csv'

# ---------- 1. 读取工具函数 -------------
def read_csv_str(path: Path) -> pd.DataFrame:
    """优先尝试 GBK，再尝试 UTF-8 读入为字符串 DataFrame."""
    for enc in ('gbk', 'utf-8', 'utf-8-sig'):
        try:
            return pd.read_csv(path, dtype=str, encoding=enc, low_memory=False)
        except UnicodeDecodeError:
            continue
    raise UnicodeDecodeError(f'无法以 GBK/UTF-8 解码 {path}')

# ------------- 2. 读取原始数据 ----------
orders = read_csv_str(FILE_ORDER)
trades = read_csv_str(FILE_TRADE)

# ------------- 3. 预处理逐笔委托 ----------
#   - 将 “方向” 列改名，避免和成交表重名冲突
orders = orders.rename(columns={'委托代码': '方向_委托'})
#   - 主键统一去掉前导 0
orders['交易所委托号'] = orders['交易所委托号'].str.lstrip('0')

# ------------- 4. 逐笔成交 → 事件流 ------
# 拆买/卖两侧，统一命名为 交易所委托号，并保留方向
buy_cols  = ['自然日', '时间', '成交代码', '成交价格', '成交数量', '叫买序号']
sell_cols = ['自然日', '时间', '成交代码', '成交价格', '成交数量', '叫卖序号']

buy  = trades[trades['叫买序号']  != '0'][buy_cols].copy()
sell = trades[trades['叫卖序号'] != '0'][sell_cols].copy()

buy .rename(columns={'叫买序号' : '交易所委托号'}, inplace=True)
sell.rename(columns={'叫卖序号': '交易所委托号'}, inplace=True)

buy ['方向_事件']  = 'B'
sell['方向_事件'] = 'S'

events = pd.concat([buy, sell], ignore_index=True)

# 事件类型: 成交(0) / 撤单(C)
events['事件类型'] = np.where(events['成交代码'] == 'C', '撤单', '成交')
events['交易所委托号'] = events['交易所委托号'].str.lstrip('0')

# ------------- 5. 合并 ------------------
merged = events.merge(
    orders,
    on='交易所委托号',
    how='left',
    suffixes=('_事件', '_委托'),
    indicator=True
)

# 合并质量检查
print("合并结果:")
print(merged['_merge'].value_counts())
# 删除指示列
merged.drop(columns=['_merge'], inplace=True)

# ------------- 6. 时间戳与存活时间 -------
fmt = '%Y%m%d%H%M%S%f'         # yyyyMMddHHmmssfff
# 委托时间
merged['委托_datetime'] = pd.to_datetime(
    merged['自然日_委托'].str.zfill(8) + merged['时间_委托'].str.zfill(9),
    errors='coerce', format=fmt
)
# 事件时间
merged['事件_datetime'] = pd.to_datetime(
    merged['自然日_事件'].str.zfill(8) + merged['时间_事件'].str.zfill(9),
    errors='coerce', format=fmt
)
merged['存活时间_ms'] = (
    merged['事件_datetime'] - merged['委托_datetime']
).dt.total_seconds() * 1_000

# ------------- 7. 去除多余列 -------------
merged = merged.loc[:, ~merged.columns.str.contains('^Unnamed')]

# ------------- 8. 保存输出 ---------------
merged.to_csv(OUTPUT_FILE, index=False, encoding='utf-8-sig')
print(f'✔️ 已生成: {OUTPUT_FILE}  ({len(merged):,} 行)')
```

```python
# 1.1 事件类型 & 方向
event_counts = df['事件类型'].value_counts()
direction_counts = df['方向_事件'].value_counts()

print('--- 事件类型分布 ---')
print(event_counts, '\n')
print('--- 方向分布 ---')
print(direction_counts)

# 1.2 委托类型
ord_type_map = {'0': 'limit', '1': 'market', 'U': 'best_own'}
df['ord_type'] = df['委托类型'].map(ord_type_map).fillna('unknown')
print('\n--- 委托类型分布 ---')
print(df['ord_type'].value_counts())

# %% [markdown]
# ## 2. 撤单率分析

# %%
cancel_rate_global = (df['事件类型'] == '撤单').mean()
print(f'整体撤单率: {cancel_rate_global:.2%}')

# 2.1 按方向
cancel_by_side = df.groupby('方向_事件')['事件类型'].apply(
    lambda x: (x == '撤单').mean()
)
print('\n--- 按方向撤单率 ---')
print(cancel_by_side)

# 2.2 按分钟
df['minute'] = df['事件_datetime'].dt.floor('T')
cancel_minute = df.groupby('minute')['事件类型'].apply(
    lambda x: (x == '撤单').mean()
)

plt.figure(figsize=(12, 4))
cancel_minute.plot()
plt.xlabel('时间')
plt.ylabel('撤单率')
plt.title('分钟级撤单率随时间变化')
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 3. 订单存活时间

# %%
# 描述统计
life_desc = df['存活时间_ms'].describe(
    percentiles=[.01, .05, .1, .25, .5, .75, .9, .95, .99]
)
print('--- 存活时间 (ms) 描述统计 ---')
print(life_desc)

# 3.1 直方图（对数纵轴）
plt.figure(figsize=(8, 4))
df['存活时间_ms'].plot.hist(bins=120, log=True)
plt.xlabel('存活时间 (ms)')
plt.ylabel('频数 (对数)')
plt.title('订单存活时间分布')
plt.tight_layout()
plt.show()

# 3.2 箱线图：不同委托类型的存活时间
plt.figure(figsize=(6, 4))
df.boxplot(column='存活时间_ms', by='ord_type', showfliers=False)
plt.yscale('log')
plt.xlabel('委托类型')
plt.ylabel('存活时间 (ms, log)')
plt.title('不同委托类型的存活时间分布')
plt.suptitle('')
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 4. 快速撤单阈值曲线

# %%
thresholds = np.arange(0, 5001, 250)  # 0–5000ms
fast_cancel_ratio = []
for th in thresholds:
    within = df['存活时间_ms'] < th
    if within.sum() == 0:
        fast_cancel_ratio.append(np.nan)
    else:
        fast_cancel_ratio.append(((df['事件类型'] == '撤单') & within).mean())

plt.figure(figsize=(8, 4))
plt.plot(thresholds, fast_cancel_ratio, marker='o')
plt.xlabel('阈值 (ms)')
plt.ylabel('阈值内撤单比例')
plt.title('快速撤单比例随阈值变化')

import textwrap

explanation = textwrap.dedent("""
    “快速撤单比例随阈值变化”的含义：
    
    我们把“快速撤单”定义为“存活时间（下单到撤单）小于某个阈值（比如 500 ms、1000 ms 等）”，
    然后对一系列不同的阈值逐个计算：在“存活时间 < 阈值”的所有委托里，有多少比例最终是被撤单的。
    
    最终绘制出来的曲线，就是当我们把“快速”这个标准从 0 ms 逐步放宽到 5000 ms 时，
    对应时间窗口里的撤单占比如何变化。
""")

print(explanation)

plt.grid(True)
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 5. 撤单热力图（时间 × 价格）
# 示例：上午 9:25–9:35 区间，价格维以 100 分为一个档。

# %%
# 过滤区间
mask_925_935 = (
    (df['事件_datetime'].dt.time >= pd.to_datetime('09:25').time()) &
    (df['事件_datetime'].dt.time <= pd.to_datetime('09:35').time())
)
df_slice = df[mask_925_935 & (df['事件类型'] == '撤单')].copy()

# 构造时间 & 价格桶
df_slice['t_bucket'] = df_slice['事件_datetime'].dt.floor('S')   # 秒级
df_slice['p_bucket'] = (df_slice['委托价格'].astype(int) // 100) * 100  # 100分档

pivot = df_slice.pivot_table(
    index='p_bucket',
    columns='t_bucket',
    values='交易所委托号',
    aggfunc='count',
    fill_value=0
)

plt.figure(figsize=(12, 6))
plt.imshow(pivot, aspect='auto')
plt.colorbar(label='撤单次数')
plt.ylabel('价格档 (分)')
plt.xlabel('时间 (秒级)')
plt.title('09:25–09:35 撤单热力图')
plt.tight_layout()
plt.show()
```

## 后验聚合

### 数据预处理（运行一次即可）

```python
# ===== ①A. 库与路径 =====
import pandas as pd
DATA_DIR = '/home/ma-user/code/fenglang/Spoofing Detect/data/exmaple'

ORDERS_CSV = f'{DATA_DIR}/委托事件流.csv'
QUOTE_CSV  = f'{DATA_DIR}/行情.csv'
BASE_OUT   = 'agg_base.parquet'          # 聚合后保存路径

# ===== ①B. 若已存在基础文件可直接加载 =====
try:
    agg_base = pd.read_parquet(BASE_OUT)
    print('√ 已加载本地缓存 agg_base.parquet')
except FileNotFoundError:
    print('→ 未找到缓存，开始重新构建 …')

    # ---------- 1. 读订单事件 ----------
    orders = pd.read_csv(
        ORDERS_CSV,
        parse_dates=['委托_datetime', '事件_datetime'],
        encoding='utf-8-sig',
        low_memory=False,
    )
    orders['委托_dt'] = pd.to_datetime(orders['委托_datetime'])
    orders['事件_dt'] = pd.to_datetime(orders['事件_datetime'])

    # ---------- 2. 聚合到“交易所委托号” ----------
    first_submit = (
        orders.sort_values('委托_dt')
              .groupby('交易所委托号', as_index=False)
              .first()[['交易所委托号', '方向_委托', '委托数量', '委托价格', '委托_dt']]
    )
    last_event = (
        orders.sort_values('事件_dt')
              .groupby('交易所委托号')['事件_dt']
              .last()
              .reset_index(name='finish_time')
    )
    exec_qty = (
        orders[orders['事件类型'] == '成交']
        .groupby('交易所委托号')['成交数量']
        .sum()
        .reset_index(name='exec_qty')
    )
    canceled = (
        orders[orders['事件类型'] == '撤单']
        .groupby('交易所委托号')
        .size()
        .reset_index(name='canceled_count')
    )
    canceled['canceled'] = 1

    agg_base = (
        first_submit
        .merge(last_event, on='交易所委托号', how='left')
        .merge(exec_qty,  on='交易所委托号', how='left')
        .merge(canceled[['交易所委托号', 'canceled']], how='left')
        .fillna({'exec_qty': 0, 'canceled': 0})
    )

    # ---------- 3. 后验特征 ----------
    agg_base['life_ms']    = (agg_base['finish_time'] - agg_base['委托_dt']).dt.total_seconds() * 1000
    agg_base['fill_ratio'] = agg_base['exec_qty'] / agg_base['委托数量']

    # ---------- 4. 关联行情 ----------
    quote = pd.read_csv(QUOTE_CSV, encoding='gbk',
                        usecols=['时间', '申买价1', '申卖价1'])
    quote['quote_dt'] = pd.to_datetime(
        quote['时间'].astype(str).str.zfill(9), format='%H%M%S%f')
    quote = quote.sort_values('quote_dt')

    agg_base['prev_quote_dt'] = agg_base['委托_dt'].apply(
        lambda t: quote.loc[quote['quote_dt'] <= t, 'quote_dt'].max()
    )
    agg_base = agg_base.merge(
        quote[['quote_dt', '申买价1', '申卖价1']],
        left_on='prev_quote_dt', right_on='quote_dt', how='left'
    )
    agg_base['spread']     = agg_base['申卖价1'] - agg_base['申买价1']
    agg_base['mid_price']  = (agg_base['申卖价1'] + agg_base['申买价1']) / 2
    agg_base['delta_mid']  = agg_base['委托价格'] - agg_base['mid_price']

    # ---------- 5. 保存 ----------
    agg_base.to_parquet(BASE_OUT, index=False)
    print(f'√ 基础特征池已保存至 {BASE_OUT}  (行数={len(agg_base)})')
```

### 调节规则 + 负样本采样（每次实验都可运行）

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
代码块 ②（升级版）：区分集合竞价 & 连续竞价，打印丢弃的集合竞价数量
-------------------------------------------------
1. 读取 agg_base.parquet
2. 按时间段标记 session
3. 统计并打印“集合竞价”行数
4. 对不同 session 施加各自规则
5. 可选择过滤 auction 样本或合并训练
6. 输出 labels_supervised.parquet
"""

import pandas as pd
from pathlib import Path
from datetime import time

BASE_OUT  = 'agg_base.parquet'
LABEL_OUT = 'labels_supervised.parquet'

# ========== ① 读取基础表 ==========
if not Path(BASE_OUT).exists():
    raise FileNotFoundError(f'{BASE_OUT} 未找到，请先运行代码块①')
agg = pd.read_parquet(BASE_OUT).copy()

# ========== ② 标记交易阶段 ==========
def tag_session(t: pd.Timestamp) -> str:
    tm = t.time()
    if time(9,15) <= tm < time(9,30):
        return 'auction_open'      # 开盘集合竞价
    if time(14,57) <= tm <= time(15,0):
        return 'auction_close'     # 收盘集合竞价
    if time(9,30) <= tm < time(11,30):
        return 'continuous_morning'
    if time(13,0) <= tm < time(14,57):
        return 'continuous_afternoon'
    return 'other'

agg['session'] = agg['委托_dt'].apply(tag_session)

# ========== ③ 统计并打印集合竞价行数 ==========
num_auction = agg['session'].str.contains('auction').sum()
print(f"共检测到 {num_auction} 条集合竞价数据，将在后续步骤中丢弃或单独处理。")

# ========== ④ 可调参数 ==========
# ---- 连续竞价规则阈值 ----
CONT_R1_MS   = 50
CONT_R2_MS   = 1000
CONT_R2_MULT = 4.0
CONT_AND     = True       # True=R1∧R2, False=R1∨R2

# ---- 集合竞价规则阈值 ----
USE_AUCTION  = False      # False=集合竞价样本全部剔除
AUCT_R1_DIFF = 0.02       # |委托价-昨收价|/昨收 ≤ 阈值（示例）
AUCT_R2_QTY  = 0.1        # 挂单量 / 当档总量 ≥ 阈值（示例）

# ---- 负样本处理 ----
BALANCE   = True
RATIO_NEG = 3

# ========== ⑤ 连续竞价标签 ==========
cont = agg[agg['session'].str.contains('continuous')].copy() 
cont['flag_R1'] = (cont['life_ms'] < CONT_R1_MS) & (cont['fill_ratio'] == 0)  
cont['flag_R2'] = (cont['life_ms'] < CONT_R2_MS) & (cont['delta_mid'].abs() >= CONT_R2_MULT * cont['spread'])
cont['y_label'] = ((cont['flag_R1'] & cont['flag_R2']) if CONT_AND              # 这两条规则一起用时，其实就是如果挂出价比较偏，且最后的确快速撤单了的话，就在主观和客观上构成了虚假报单
                   else (cont['flag_R1'] | cont['flag_R2'])).astype(int)

# ========== ⑥ 集合竞价标签（可选） ==========
if USE_AUCTION:
    auc = agg[agg['session'].str.contains('auction')].copy()
    # 示例规则：距离昨收价过大 + 挂单量占比异常大
    auc['flag_A1'] = (auc['delta_mid'].abs() / auc['mid_price'].abs()) > AUCT_R1_DIFF
    auc['flag_A2'] = (auc['委托数量'] / auc.groupby('委托价格')['委托数量'].transform('sum')) > AUCT_R2_QTY
    auc['y_label'] = (auc['flag_A1'] & auc['flag_A2']).astype(int)
    dataset = pd.concat([cont, auc], ignore_index=True)
else:
    # 直接丢弃集合竞价样本
    print(f"丢弃了 {num_auction} 条集合竞价数据，仅使用连续竞价进行标签构造。")
    dataset = cont.copy()

# ========== ⑦ 负样本抽样 ==========
pos = dataset[dataset['y_label'] == 1]
neg_pool = dataset[dataset['y_label'] == 0]

if BALANCE:
    n_neg = min(len(pos) * RATIO_NEG, len(neg_pool))
    neg = neg_pool.sample(n=n_neg, random_state=42)
else:
    neg = neg_pool

train_df = pd.concat([pos, neg], ignore_index=True)

# ========== ⑧ 保存 ==========
train_df[['交易所委托号', 'y_label']].to_parquet(LABEL_OUT, index=False)

print(f"连续竞价行数: {len(cont)} | 集合竞价启用: {USE_AUCTION}")
print(f"最终正样本: {len(pos)} | 负样本: {len(neg)} (比例 ≈ {len(pos)}:{len(neg)})")
print(f"√ 标签文件已保存 ➜ {LABEL_OUT}")
```

```python
#!/usr/bin/env python
# build_features_from_raw.py
# -------------------------------------------------------
# 读取委托事件流 + 行情 CSV  👉  实时可观测特征 X_features.parquet
# 依赖：pandas≥2.0  numpy

import pandas as pd, numpy as np
from pathlib import Path

# 1. 路径
ORDERS_CSV = '/home/ma-user/code/fenglang/Spoofing Detect/data/exmaple/委托事件流.csv'
QUOTE_CSV  = '/home/ma-user/code/fenglang/Spoofing Detect/data/exmaple/行情.csv'
FEAT_OUT   = 'X_features.parquet'

# 2. 读取逐笔委托事件流 -------------------------------------------------
orders = pd.read_csv(
    ORDERS_CSV,
    encoding='utf-8-sig',
    parse_dates=['委托_datetime', '事件_datetime'],
    low_memory=False
)

# 3. 读取 L1 行情，只要买一卖一价 --------------------------------------
quotes = pd.read_csv(
    QUOTE_CSV,
    encoding='gbk',
    usecols=['自然日', '时间', '申买价1', '申卖价1', '前收盘']
)

# 3.1 把 '时间' (HHMMSSmmm) 转成 Timestamp（缺日期 → 用 '自然日' 补）
quotes['quote_dt'] = pd.to_datetime(
    quotes['自然日'].astype(str) + quotes['时间'].astype(str).str.zfill(9),
    format='%Y%m%d%H%M%S%f'
)
quotes = quotes.sort_values('quote_dt')

# 4. 把 L1 行情 merge_asof 到订单下单时刻 ------------------------------
orders = orders.sort_values('委托_datetime')
orders = pd.merge_asof(
    orders, quotes[['quote_dt', '申买价1', '申卖价1', '前收盘']],
    left_on='委托_datetime', right_on='quote_dt',
    direction='backward'      # 最近 ≤t 的快照
)

# 5. 构造实时可观测特征 -----------------------------------------------
df = orders.copy()

# 5.1 方向
df['is_buy'] = (df['方向_委托'] == '买').astype(int)

# 5.2 中间价、价差、相对偏移
df['mid_price']  = (df['申买价1'] + df['申卖价1']) / 2
df['spread']     =  df['申卖价1'] - df['申买价1']
df['delta_mid']  =  df['委托价格'] - df['mid_price']
df['pct_spread'] =  (df['delta_mid'].abs() / df['spread']).replace([np.inf, -np.inf], 0).fillna(0)

# 5.3 订单规模
df['log_qty'] = np.log1p(df['委托数量'])

# 5.4 订单流密度：100 ms 内的订单数  (必须按时间排序)
df = df.sort_values('委托_datetime')
df['orders_100ms'] = (
    df.rolling('100ms', on='委托_datetime')['委托_datetime']
      .count()
      .shift(1)
      .fillna(0)
)

# 5.5 撤单活跃度：过去 5 s 内撤单次数
df['cancel_flag'] = (df['事件类型'] == '撤单').astype(int)
df['cancels_5s'] = (
    df.rolling('5s', on='委托_datetime')['cancel_flag']
      .sum()
      .shift(1)
      .fillna(0)
)

# 5.6 日内周期特征 (sin / cos)
sec_since_open = (
    df['委托_datetime'] - df['委托_datetime'].dt.normalize() - pd.Timedelta('09:30:00')
).dt.total_seconds()
day_seconds = 4.5 * 3600
df['time_sin'] = np.sin(2*np.pi * sec_since_open / day_seconds)
df['time_cos'] = np.cos(2*np.pi * sec_since_open / day_seconds)

# 5.7 集合竞价标记 + 相对昨收偏离
def in_auction(ts):
    tm = ts.time()
    return 1 if (pd.Timestamp('09:15').time() <= tm < pd.Timestamp('09:30').time()
                 or pd.Timestamp('14:57').time() <= tm <= pd.Timestamp('15:00').time()) else 0

df['in_auction'] = df['委托_datetime'].apply(in_auction)
df['price_dev_prevclose'] = (df['委托价格'] - df['前收盘']) / df['前收盘']

# 6. 选列并保存 ---------------------------------------------------------
features = [
    '交易所委托号',
    'is_buy', 'delta_mid', 'pct_spread',
    'orders_100ms', 'cancels_5s',
    'log_qty', 'time_sin', 'time_cos',
    'in_auction', 'price_dev_prevclose'
]
df[features].fillna(0).to_parquet(FEAT_OUT, index=False)
print(f'√ 特征文件已保存 ➜ {FEAT_OUT}  (样本 {len(df)})')
```
